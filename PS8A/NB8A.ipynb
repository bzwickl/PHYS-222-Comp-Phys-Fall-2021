It is often nice to **normalize** the counts so the y-axis is given as a **fraction** (i.e., a number between 0 and 1) that can be compared with the expected probability. You should create a plot something like this:

![Fraction with error bar](Fraction_with_error_bar.png)

1. Although we expect each side to come up 1/6 of the time, in any particular set of rolls we may get more or less. There is a formula to compute this uncertainty, often called the standard error, which is given by $\sigma = \frac{\Delta n}{N} = \frac{\sqrt{p(1-p)N}}{N}$.  The deviation from $N/6$ should typically fall within $\Delta n$, which is the theoretical standard deviation (or error) in how many times a particular outcome happens.  $p$ is the expected probability of the outcome. $N$ is the total number of rolls/trials.  Your error bars should have a magnitude of $\pm \sigma$. 
1. Use the `plt.errorbar()` function to add an error bar to the **expected value**. See the [official Matplotlib documentation](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.errorbar.html?highlight=errorbar#matplotlib.pyplot.errorbar) for inputs to `errorbar()`.

The function `np.full(shape,fill_value)` returns a new array of a given shape (or length) filled with *fill_value*. Both `np.ones` and `np.full` are useful for quickly generating an array of constant values. See documentation for [np.full()](https://numpy.org/doc/stable/reference/generated/numpy.full.html#numpy.full) and [np.ones](https://numpy.org/doc/stable/reference/generated/numpy.ones.html#numpy.ones).

# Task 3: Simulating two dice rolls
Using a minimal modification of Task 1, model the sum of a pair of dice and plot the distribution. 

Function name: `two_dice(Nroll)`

Inputs:  
* `Nroll` a positive integer representing the number of times both dice are rolled

Outputs: 
* Display a bar graph showing the distribution of results for the sum of the two dice.
* returns a list `[unique_sum, unique_sum_counts]`, where
    * `unique_sum` a numpy array containing the unique values the sum takes on
    * `unique_sum_counts`  the number of counts each value of the sum occurs

```python
""" Calculcate random arrays representing dice rolls """

### BEGIN SOLUTION
def two_dice(Nroll) :
    y1 = np.random.randint(1,7,Nroll)
    y2 = np.random.randint(1,7,Nroll)
    unique, unique_counts = np.unique(y1 + y2,return_counts=True)
    plt.figure()
    plt.bar(unique, unique_counts)
    plt.xlabel("Dice sum value")
    plt.ylabel("Number of occurances")
    plt.title("Number of rolls = " + str(Nroll))
    return [unique, unique_counts]
    
### END SOLUTION
```

```python
""" Test cases for the two_dice() function """

Nroll = 100
unique, unique_counts = two_dice(Nroll)
assert np.sum(unique_counts) == Nroll
assert min(unique) >= 2
assert max(unique) <= 12
print(unique_counts)
print(Nroll/6)
print("Theoretical Error (rolling a 7) = ", np.sqrt(Nroll * 1/6 * 5/6))
print("Actual error (rolling a 7) = ", abs(unique_counts[5] - Nroll/6))
assert abs(unique_counts[5] - Nroll/6) < 3*np.sqrt(Nroll * 1/6 * 5/6)

Nroll = 10000
unique, unique_counts = two_dice(Nroll)
assert np.sum(unique_counts) == Nroll
assert min(unique) == 2
assert max(unique) == 12
print(unique_counts)
print(Nroll/6)
print("Theoretical Error (rolling a 7) = ", np.sqrt(Nroll * 1/6 * 5/6))
print("Actual error (rolling a 7) = ", abs(unique_counts[5] - Nroll/6))
assert abs(unique_counts[5] - Nroll/6) < 3*np.sqrt(Nroll * 1/6 * 5/6)
```

# Using `cumsum()` (cumulative sum) to generate a random walk

As discussed in lecture, a **[random walk](https://en.wikipedia.org/wiki/Random_walk)** is a random process that starts at 0 and with each step moves +1 or -1 with equal probability.

### Example of a single 1D random walk with $N_\textrm{steps}$ steps.


```python
""" Example code for a 1D random walk with Nsteps """
Nsteps = 20

y = np.random.randint(0, 2, Nsteps)  # array of 0,1 values
print("y = ", y)

y1 = (-1)**y    # converts array of 0s & 1s to +1s & -1s.
print("y1 = ", y1)

walk = y1.cumsum()  # apply cumulative sum to generate walk
print("walk = ", walk )

x = np.arange(1,Nsteps+1)  # array of step numbers 1 - Nsteps
plt.plot(x, walk)  # plot position of walker vs step number
plt.xlabel("Step number")
plt.ylabel("Position")
```

```python
""" Calculate 100 objects going on random walks with 1000 steps"""

### BEGIN SOLUTION
Nsteps = 1000
Nwalkers = 100

y = np.random.randint(0, 2, (Nsteps,Nwalkers))  # array of 0,1 values
y1 = (-1)**y    # converts array of 0s & 1s to +1s & -1s.
print(y1.shape)

walk = y1.cumsum(axis=0)
walk_mean = walk.mean(axis=1)
walk_std = walk.std(axis=1)

#plt.plot(walk)
x = np.arange(1,Nsteps+1)
plt.plot(x,walk_mean);
plt.xlabel("Step number")
plt.ylabel("Mean position of the "+str(Nwalkers)+" objects")
plt.title("The mean position stays close to zero")
#plt.savefig("Random_walk_mean_position.png")

plt.figure()
plt.xlabel("Step number")
plt.ylabel("Standard deviation of positions ")
plt.title("Standard deviation grows like $\sqrt{N_{steps}}$")


plt.plot(x,walk_std);
plt.plot(x,x**0.5)

#plt.savefig("Random_walk_std_position.png")
### END SOLUTION
```

# Comment on random walks and diffusion
1. Brownian motion, which can be modeled as a random walk, is an example of diffusive motion. In regular 1D kinematics, objects move with a well defined velocity that causes the position to change linearly with time ($ x = v t$). In diffusive 1D kinematics, objects randomly move, however their expected net displacement increases like $ \Delta x \propto \sqrt{t}$.
1. A large number of fictional particles is often called an **ensemble**, and we study their dynamics using statistical quantities like `mean` and standard deviation.
1. Sometimes a model that uses an ensemble of randomly varying outcomes is called a **Monte Carlo simulation**, named after the [region in country Monaco](https://en.wikipedia.org/wiki/Monte_Carlo) with lots casinos and gambling.
